```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

import tensorflow
from numpy.random import seed
np.random.seed(1)
PYTHONHASHSEED=0
tensorflow.random.set_seed(2)


columns=["Class", "age", "menopause", "tumor-size", "inv-nodes", "node-caps", "deg-malig", "breast", "breast-quad", "irradiat"]
cancer = pd.read_csv('data/breast-cancer.data', header=None, names =columns)


cancer['Target'] = cancer['Class'].map(lambda x: 1 if x=='recurrence-events' else 0 )
cancer.drop("Class", axis=1,inplace=True)

target = cancer[['Target']]
cancer.drop('Target', axis = 1, inplace = True)




# Question 1

def modifyAge():
    def changeAge(x):
        if x=='20-29':
            return 1
        elif x=='30-39':
            return 2
        elif x=='40-49':
            return 3
        elif x=='50-59':
            return 4
        elif x=='60-69':
            return 5
        elif x=='70-79':
            return 6
    
    cancer['age'] = cancer['age'].apply(lambda x: changeAge(x))
    return cancer['age']

modifyAge()


def changeTumor(x):
    if x == '0-4':
        return 1
    elif x == '5-9':
        return 2
    elif x == '10-14':
        return 3
    elif x == '15-19':
        return 4
    elif x == '20-24':
        return 5
    elif x == '25-29':
        return 6
    elif x == '30-34':
        return 7
    elif x == '35-39':
        return 8
    elif x == '40-44':
        return 9
    elif x == '45-49':
        return 10
    elif x == '50-54':
        return 11

cancer['tumor-size']=cancer['tumor-size'].apply(lambda x: changeTumor(x))

def changeInv(x):
    if x == '0-2':
        return 1
    elif x == '3-5':
        return 2
    elif x == '6-8':
        return 3
    elif x == '9-11':
        return 4
    elif x == '12-14':
        return 5
    elif x == '15-17':
        return 6
    elif x == '24-26':
        return 7

cancer['inv-nodes'] = cancer['inv-nodes'].apply(lambda x: changeInv(x))

cancer_dummies = pd.get_dummies(cancer, drop_first=True)

new_df = cancer_dummies

# Question 2

def question2():
    X_train, X_test, y_train, y_test = train_test_split(new_df, target, test_size=0.3, random_state = 101)
    return X_train, X_test, y_train, y_test


X_train, X_test, y_train, y_test = question2()


# Question 3
from sklearn.preprocessing import StandardScaler

def question3():
    sc = StandardScaler()
    X_train_scaled = sc.fit_transform(X_train)
    X_test_scaled = sc.transform(X_test)
    return X_train_scaled, X_test_scaled


# Question 3
X_train_scaled, X_test_scaled = question_3()

#Question 4
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

def question_4():
    classifier = Sequential()
    classifier.add(Dense(units=32,kernel_initializer = 'uniform', activation= 'relu', input_dim = 15))
    classifier.add(Dense(units=32, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units=32, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units=32, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units= 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

    classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    classifier.fit(X_train_scaled, y_train, batch_size=50, epochs = 20)

    loss, accuracy = classifier.evaluate(X_test_scaled, y_test)

    return loss, accuracy

loss, accuracy = question4()
print(loss)
print(accuracy)

# Question 5
def question5():

    classifier2 = Sequential()
    classifier2.add(Dense(units=32, kernel_initializer = 'uniform', activation = 'relu',input_dim=15))
    classifier2.add(Dropout(0.3))

    classifier2.add(Dense(units=32, kernel_initializer = 'uniform', activation = 'relu'))
    classifier2.add(Dropout(0.3))

    classifier2.add(Dense(units=32, kernel_initializer = 'uniform', activation = 'relu'))
    classifier2.add(Dropout(0.3))

    classifier2.add(Dense(units=32, kernel_initializer = 'uniform', activation = 'relu'))
    classifier2.add(Dropout(0.3))

    classifier2.add(Dense(units= 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

    classifier2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    classifier2.fit(X_train_scaled, y_train, batch_size=50, epochs = 20)
    loss, accuracy= classifier2.evaluate(X_test_scaled, y_test)

    return loss, accuracy

loss2, accuracy2 = question5()
print()
print(loss2)
print(accuracy2)
```
